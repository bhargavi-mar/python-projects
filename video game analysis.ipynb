# how many games have been sold between the year 2000 and 2010 globally?


```python
year_count = [i for i in range(2000,2011)]
count_in_range = data.loc[data['Year'].isin(year_count)]
```

```python
count_in_range
```

```python
north_america_sales = sum(count_in_range.NA_Sales)
print('\n Total sales in north america from 2001 to 2010 is:', north_america_sales)

europe_sales = sum(count_in_range.EU_Sales)
print('\n Total sales in Europe from 2001 to 2010 is:', europe_sales)

Japan_sales = sum(count_in_range.JP_Sales)
print('\n Total sales in Japan from 2001 to 2010 is:', Japan_sales)

global_sales = sum(count_in_range.Global_Sales)
print('\n Total sales in globally from 2001 to 2010 is:', global_sales)
```

# find the average , min ,max global sales based on publisher 

```python
data.groupby('Publisher')['Global_Sales'].mean()
```

```python
data.groupby('Publisher')['Global_Sales'].min()
```

```python
data.groupby('Publisher')['Global_Sales'].max()
```

```python
aggregate_result = data.groupby('Publisher').agg({'Global_Sales' : ['mean','min','max']})
aggregate_result
```

# how are the racing game performing over the time??

```python
racing_data = data.loc[data['Genre'] =='Racing']
racing_data
```

```python
plt.figure(figsize =(24,12))
sns.barplot(x='Year',y='Global_Sales', data= racing_data)
plt.title('Global sale year wise for genre Racing')
plt.show()
```
