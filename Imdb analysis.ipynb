# classify Movie based on Rating[excellent, good and average]

```python
def rating(rating):
    if rating >=7.0:
        return "Excellent"
    elif rating >=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['rating_category']= data['Rating'].apply(rating)
data.head()
```

data.drop('rating_cat', axis=1, inplace= True)

```python
list1=[]
for values in data['Genre']:
    list1.append(values.split(','))
    
```

```python
list1
```

```python
# convert the 2d list into 1d list 
one_d =[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
unique_list =[]
for item in one_d:
    if item not in unique_list:
        unique_list.append(item)
```
